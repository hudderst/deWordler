```python
#ingest words

words = []

files = ['data.adj','data.adv','data.noun','data.verb']

for file in files:
    with open(file, 'r') as f:        
        for l in f.readlines():
            wrds = l.split(' ')
            for w in wrds:
                if len([a for a in w if a.isalpha()]) == 5:
                    words.append(''.join([a.lower() for a in w if a.isalpha()]))
            
                
```
